```python
import numpy as np
import math
import matplotlib.pyplot as plt
dataPoints = 500

mean = [0,0]
covariance1 = [[1,0],[0,1]]
covariance2 = [[0.2,0],[0,0.2]]
covariance3 = [[2,0],[0,2]]
covariance4 = [[0.3,0.5],[0.5,2]]
data1 = np.random.multivariate_normal(mean,covariance1,dataPoints)
data2 = np.random.multivariate_normal(mean,covariance2,dataPoints)
data3 = np.random.multivariate_normal(mean,covariance3,dataPoints)
data4 = np.random.multivariate_normal(mean,covariance4,dataPoints)
```

```python
plt.scatter(data1[:, 0], data1[:, 1], marker='.', alpha=0.5, color='red')
plt.title('Set 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
plt.scatter(data2[:, 0], data2[:, 1], marker='.', alpha=0.5, color='red')
plt.title('Set 2')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
plt.scatter(data3[:, 0], data3[:, 1], marker='.', alpha=0.5,color='red')
plt.title('Set 3')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
plt.scatter(data4[:, 0], data4[:, 1], marker='.', alpha=0.5, color='red')
plt.title('Set 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
```
