```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_normal_dist(mu,sigma,num_pts):
    x=np.linspace(mu-3*sigma,mu+3*sigma,num_pts)
    pdf=(1/(sigma*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)/sigma)**2)
    plt.plot(x,pdf,label="normal dist")
    plt.title("Normal Distribution")
    plt.xlabel('x')
    plt.plot('prob density')
    plt.grid()

plot_normal_dist(3,50,10000)
```

```python
def plot_exp_dist(la,num_pts):
    x=np.linspace(0,10,100)
    pdf=(la*np.exp(-1*la*x))
    plt.plot(x,pdf,label="exp dist")
    plt.title("exponential Distribution")
    plt.xlabel('x')
    plt.plot('prob density')
    plt.grid()

plot_exp_dist(1,1000)
```

```python
from scipy.special import factorial
def plot_exp_dist(la,num_pts):
    t = np.arange(0, 20, 0.1)
    d = np.exp(-1*la)*np.power(la, t)/factorial(t)
    plt.plot(t, d)
    plt.title("Poisson's Distribution")
    plt.grid()
    
plot_exp_dist(4,100)
```

```python
def plot_cauchy_dist(num_pts):
    x=np.linspace(-10,10,100)
    pdf=1/((np.pi)*(1 + x*x))
    plt.plot(x,pdf,label="chy dist")
    plt.title("Cauchy Distribution")
    plt.xlabel('x')
    plt.plot('prob density')
    plt.grid()
    
plot_cauchy_dist(10)
```

```python
import math
def plot_chiSq_dist(k, num_pts):
    x=np.linspace(0,50,1000)
    pdf=(np.power(x,(k/2)-1)*np.exp(-0.5*x))/np.power(2,k/2)*math.gamma(k/2)
    plt.plot(x,pdf,label="chi-sq dist")
    plt.title("Chi squared Distribution")
    plt.xlabel('x')
    plt.plot('prob density')
    plt.grid()
    
plot_chiSq_dist(20,1000)
```

```python
import math 
def plot_gamma_dist(theta, k, num_pts):
    x=np.linspace(0,10,1000)
    pdf=((x ** (k - 1)) * np.exp(-x*theta)*(theta**x))/math.gamma(k)
    plt.plot(x,pdf,label="Gamma dist")
    plt.title("Gamma Distribution")
    plt.xlabel('x')
    plt.plot('prob density')
    plt.grid()
    
plot_gamma_dist(15,30,1000)
```
